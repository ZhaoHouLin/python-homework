# 作業 #03：在靜態網站的頁面中取得更多的資訊

## 題目描述
到目前為主，我們學習到從靜態頁面中可以取得「一個頁面」與「多個頁面」的資料。接下來請你將「作業 #02：在靜態網站的頁面中取得更多的資訊」的結果模仿批次自動化資料收集，試著取回更多商品的資訊並儲存起來。

```python
import requests
from bs4 import BeautifulSoup

headers = {
  'user-agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.108 Safari/537.36'  
}


url = 'https://m.momoshop.com.tw/search.momo?_advFirst=N&_advCp=N&curPage=1&searchType=&cateLevel=-1&ent=k&searchKeyword=PS5&_advThreeHours=N&_isFuzzy=0&_imgSH=fourCardType'
r = requests.get(url, headers = headers)
response = r.text
soup = BeautifulSoup(r.text)
print(soup.prettify())
```

```python
items = soup.find_all('li',class_='goodsItemLi')
links = []

for item in items:
#     print(item.find('h3',class_='prdName').text.strip())
#     print(item.find('b',class_='price').text.strip())
    links.append(item.a['href'])
    
links
```

```python
products = []

for link in links:
    
    link = f"https://www.momoshop.com.tw/goods/GoodsDetail.jsp?{link.split('?')[1]}"
    
    r = requests.get(link, headers=headers)
    response = r.text
    soup = BeautifulSoup(r.text)

    product = {}
    product['網址'] = link
    product['商品名稱'] = soup.find('p', class_='fprdTitle').text
    product['價錢'] = soup.find('ul', class_='prdPrice').find('li', class_='special').span.text
    products.append(product)
    
products
```

```python
headers = {
  'user-agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.108 Safari/537.36'  
}

page = 1
keyword = 'ps5'

url = f'https://m.momoshop.com.tw/search.momo?_advFirst=N&_advCp=N&curPage={page}&searchType=&cateLevel=-1&ent=k&searchKeyword={keyword}&_advThreeHours=N&_isFuzzy=0&_imgSH=fourCardType'
r = requests.get(url, headers=headers)
response = r.text
soup = BeautifulSoup(r.text)

links = []

for item in soup.find_all('li', class_="goodsItemLi"):
    links.append(item.a['href'])

print(len(links))
links
```

```python
headers = {
  'user-agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.108 Safari/537.36'  
}

pages = 5
keyword = 'ps5'
links = []

for page in range(1, pages+1):

    url = f'https://m.momoshop.com.tw/search.momo?_advFirst=N&_advCp=N&curPage={page}&searchType=&cateLevel=-1&ent=k&searchKeyword={keyword}&_advThreeHours=N&_isFuzzy=0&_imgSH=fourCardType'
    r = requests.get(url, headers=headers)
    response = r.text
    soup = BeautifulSoup(r.text)

    for item in soup.find_all('li', class_="goodsItemLi"):
        links.append(item.a['href'])

print(len(links))
links
```
