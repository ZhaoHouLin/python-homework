# 作業 #02：在靜態網站的頁面中取得更多的資訊

## 題目描述
在前兩個實作章節中，我們成功爬取了「Yahoo 商城」和「Momo 電商」中的商品名稱、商品價格和商品網址資訊。在這個作業想請你試試看抓取不同的商品並且取得更多的資料，可以從兩個電商的商品頁取得「配送方式」、「商品特色」和「商品照片網址」。

```python
import requests
from bs4 import BeautifulSoup

headers = {
  'user-agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.108 Safari/537.36'  
}


url = 'https://www.momoshop.com.tw/goods/GoodsDetail.jsp?i_code=8267514&Area=search&mdiv=403&oid=1_1&cid=index&kw=ps5'
r = requests.get(url, headers = headers)
response = r.text
soup = BeautifulSoup(r.text)
print(soup.prettify())
```

```python
p = soup.find('p',id='osmGoodsName')
print(p.text)
```

```python
price = soup.find('ul', class_='prdPrice').find('li', class_='special').span
price = float(price.text.replace(',', ''))
price
```

### Q1: 配送方式

```python
delivery = soup.find('dl',class_='shippingList').find_all('a')
for dv in delivery:
    print(dv.text)
```

### Q2: 商品特色

```python
feature = soup.find('a',id='openvideoBtn')['vlink']
print(feature)
```
