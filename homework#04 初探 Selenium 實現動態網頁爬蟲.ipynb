## 批次收集

```python
browser = webdriver.Chrome(executable_path='./chromedriver')

browser.get("https://www.momoshop.com.tw/search/searchShop.jsp?keyword=PS5&searchType=1&curPage=1&_isFuzzy=0&showType=chessboardType")

import time
time.sleep(5)

source = browser.page_source
soup = BeautifulSoup(source)
```

```python
for d in soup.find_all(class_='goodsUrl'):
    print(d['href'])
```

```python
# 至底加載
for y in range(0, 10000, 500):
    browser.execute_script(f"window.scrollTo(0, {y})")
    time.sleep(0.5)
```

```python
source = browser.page_source
soup = BeautifulSoup(source)
```

```python
links = []

for d in soup.find_all(class_='goodsUrl'):
    links.append(d['href'])
    
links
```

```python
products = []

for link in links:
    
    link = f"https://www.momoshop.com.tw/goods/GoodsDetail.jsp?{link.split('?')[1]}"
    browser.get(link)
    time.sleep(10)  #等10秒才跳轉下一頁
    soup = BeautifulSoup(browser.page_source)

    product = {}
    product['網址'] = link
    product['商品名稱'] = soup.find('p', class_='fprdTitle').text
    product['價錢'] = soup.find('ul', class_='prdPrice').find('li', class_='special').span.text
    products.append(product)
    
products
```

## 頁面載入完自動換頁

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
products = []

for link in links[:5]:
    
    link = f"https://www.momoshop.com.tw/goods/GoodsDetail.jsp?{link.split('?')[1]}"
    browser.get(link)
    WebDriverWait(browser, 10).until(
        EC.visibility_of_element_located(
            (By.CLASS_NAME, "fprdTitle")
        )
    )

    soup = BeautifulSoup(browser.page_source)
    
    product = {}
    product['網址'] = link
    product['商品名稱'] = soup.find('p', class_='fprdTitle').text
    product['價錢'] = soup.find('ul', class_='prdPrice').find('li', class_='special').span.text
    products.append(product)
    
products
```

## 從「電商網站的動態網站爬蟲實作 - 蝦皮購物」範例中延伸出來，試著抓取更多的商品資訊：運送方式、商品特色及商品照片網址。

```python
from selenium import webdriver

browser = webdriver.Chrome(executable_path='./chromedriver')

browser.get("https://shopee.tw/search?keyword=iphone12%20128g")

import time
time.sleep(5)

source = browser.page_source
soup = BeautifulSoup(source)
```

```python
source = browser.page_source
soup = BeautifulSoup(source)
```

```python
links = []

for d in soup.find_all(class_='shopee-search-item-result__item'):
    links.append(d.a['href'])
    
links
```

```python
products = []

for d in links[:5]:
    link = f"https://shopee.tw/{d}"
    browser.get(link)
    soup = BeautifulSoup(browser.page_source)
    
    product = {}
    product['網址'] = link
    product['商品名稱'] = soup.find('div', class_='product-briefing') \
        .find('div', class_='attM6y') \
        .find('span').text
    product['價錢'] = soup.find('div', class_='_3e_UQT').text
    if soup.find('div', class_='_2dyNDF'):
        product['優惠'] = soup.find('div', class_='_2dyNDF').text
    products.append(product)
    
products
```
