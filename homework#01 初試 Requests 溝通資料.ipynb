## 題目描述
我們使用了「Requests」模擬 HTTP 請求並且攔截回應，算是完成了網頁爬蟲的取回資料的第一步。再進一步取出資料之前，想請大家試試看對以下幾個網站（網址）發出請求，並且判斷 「① 是否可以正確回傳結果？」 與 「② 回傳的結果跟網頁內容一致嗎？」。

### Q1: https://www.104.com.tw/cust/list/index/
#### ① 是
#### ② 一致

```python
import requests
import json
print(requests)
print(requests.__version__)
```

```python
r1 = requests.get('https://www.104.com.tw/cust/list/index/')
response1 = r1.text
print(response1)
```

### Q2: https://www.dcard.tw/f
#### ① 是
#### ② 一致

```python
r2 = requests.get('https://www.dcard.tw/f')
response2 = r2.text
print(response2)
```

### Q3: https://www.zhihu.com/explore
#### ① 否
#### ② 不一致

```python
r3 = requests.get('https://www.zhihu.com/explore')
response3 = r3.text
print(response3)
```

### 加 Headers 後
#### ① 是
#### ② 一致

```python
url3 = 'https://www.zhihu.com/explore'
headers = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.103 Safari/537.36'
}

r3 = requests.get(url3, headers=headers)
r3.encoding = 'utf-8'

print(r3.text)
```
